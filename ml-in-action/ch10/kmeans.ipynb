```python
from numpy import *
import matplotlib
import matplotlib.pyplot as plt

def loadDataSet(fileName):
    dataMat = []
    fr = open(fileName)
    for line in fr.readlines():
        curLine = line.strip().split('\t')
        fltLine = map(float,curLine)
        dataMat.append(fltLine)
    return array(dataMat)

def distEuclid(vecA, vecB):
    return sqrt(sum(power(vecA - vecB, 2)))

def randCent(dataSet, k):
    n = shape(dataSet)[1]
    centroids = mat(zeros((k,n)))
    for j in range(n):
        minJ = min(dataSet[:,j])
        rangeJ = float(max(dataSet[:,j]) - minJ)
        centroids[:,j] = minJ + rangeJ * random.rand(k,1)
    return centroids

datMat = loadDataSet('data/testSet.txt')
datMat[:,0]
```

```python
randCentroids = randCent(datMat, 4)

fig = plt.figure()
plt.title("kmeans with random centroids")
ax = fig.add_subplot(111)
ax.scatter(datMat[:,0], datMat[:,1])
for centroid in randCentroids:
    ax.scatter(centroid[0,0],centroid[0,1],color='r', marker='+')

plt.show()
```

```python
def kMeans(dataSet, k, distMeas=distEuclid, createCent=randCent):
    m = shape(dataSet)[0]
    clusterAssment = mat(zeros((m,2)))
    centroids = createCent(dataSet, k)
    clusterChanged = True
    while clusterChanged:
        clusterChanged = False
        for i in range(m):
            minDist = inf; minIndex = -1
            for j in range(k):
                distJI = distMeas(centroids[j,:],dataSet[i,:])
                if distJI < minDist:
                    minDist = distJI; minIndex = j
            if clusterAssment[i,0] != minIndex: clusterChanged = True
            clusterAssment[i,:] = minIndex,minDist**2
        for cent in range(k):
            ptsInClust = dataSet[nonzero(clusterAssment[:,0].A==cent)[0]]
            centroids[cent,:] = mean(ptsInClust, axis=0)
    return centroids, clusterAssment

# run kmeans 1000 times with random centroids
allCentroids = []
for i in range(1000):
    centroids, clusterAssment = kMeans(datMat,4)
    allCentroids.append(centroids)
fig = plt.figure()
plt.title("kmeans 1000 times with random centroids")
ax = fig.add_subplot(111)
ax.scatter(datMat[:,0], datMat[:,1])
for centroid in allCentroids:
    ax.scatter(centroid[0,0],centroid[0,1],color='r', marker='+')
```

```python
def kplusplusCent(dataSet, k):
    n = shape(dataSet)[1]
    m = shape(dataSet)[0]
    
    firstCentroid = randCent(dataSet, 1)
    
    centroids = mat(zeros((k,n)))
    centroids[0,:] = firstCentroid
    selectedDataPoints = []
    
    for i in range(1,k):#find ith centroid
        maxDistance = 0
        maxDataPoint = None
        remainingDataPoints = (j for j in range(m) if not j in selectedDataPoints)
        for j in remainingDataPoints: #go over all remaining datapoints
            for k in range(i): #find most distant centroid
                centroid = centroids[k,:]
                distance = distEuclid(dataSet[j,:], centroid)
                if distance > maxDistance:
                    maxDistance = distance
                    maxDataPoint = j
        selectedDataPoints.append(maxDataPoint)
        centroids[i,:] = dataSet[maxDataPoint,:]
    return centroids

centroids = kplusplusCent(datMat,4)
fig = plt.figure()
plt.title("kmeans with k++ centroids")
ax = fig.add_subplot(111)
ax.scatter(datMat[:,0], datMat[:,1])
for centroid in centroids:
    print centroid
    ax.scatter(centroid[0,0],centroid[0,1],color='r', marker='+')
```

```python
# run kmeans 1000 times with k++ centroids
allCentroids = []
for i in range(1000):
    centroids, clusterAssment = kMeans(datMat,4,createCent=kplusplusCent)
    allCentroids.append(centroids)

fig = plt.figure()
plt.title("kmeans 1000 times with k++ centroids")
ax = fig.add_subplot(111)
ax.scatter(datMat[:,0], datMat[:,1])
for centroid in allCentroids:
#     print centroid
    ax.scatter(centroid[0,0],centroid[0,1],color='r', marker='+')
```
